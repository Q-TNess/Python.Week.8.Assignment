# **TASK 3**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# STREAMS DATA OF BEYONCE'S VIDEO OVER THE YEARS 
years = [2011,2012,2015,2020,2025]
streams = [1100000,2000000,2000100,6200000,6300000]

# Line graph
plt.figure(figsize=(8,5))
sns.lineplot(x=years, y=streams, marker='o', color='purple')
plt.title("Video Streams over the Years (Line Chart)")
plt.xlabel("Year")
plt.ylabel("Streams")
plt.tight_layout()
plt.show()
```

```python
# Bar graph
plt.figure(figsize=(8,5))
sns.barplot(x=years, y=streams, palette="Blues_d")
plt.title("Video Streams over the Years (Bar Chart)")
plt.xlabel("Year")
plt.ylabel("Streams")
plt.tight_layout()
plt.show()
```

```python
# Histogram

plt.figure(figsize=(8,5))
plt.hist(streams, bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Video Streams (Histogram)')
plt.xlabel('Number of Streams')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(6,6))
plt.pie(streams, labels=years, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Video Streams over the Years (Pie Chart)")
plt.axis('equal')
plt.show()
```

### Error handling:

```python
# Attempt to load data 
try:
    import csv

    years = []
    streams = []

    with open("beyonce_streams.csv", mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            try:
                year = int(row['year'])
                stream = int(row['streams'])
                years.append(year)
                streams.append(stream)
            except ValueError:
                print(f"Invalid data found and skipped: {row}")
except FileNotFoundError:
    print("File not found. Using default data.")
```
